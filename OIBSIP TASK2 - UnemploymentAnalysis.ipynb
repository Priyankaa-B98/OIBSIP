# UNEMPLOYMENT ANALYSIS WITH PYTHON

### Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19, so analyzing the unemployment rate can be a good data science project. 

## 1. Importing the Library:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

## 2.Reading the Dataset:

```python
df2 = pd.read_csv("Unemployment in India.csv")
```

```python
df2.head()
```

## 3. Exploratory Data Analysis:

```python
df2.describe()
```

```python
df2.info()
```

- There are totally 7 columns of which 4 are of "object" datatype, columns are Region, Date, Frequency, Area.

- Also there are 3 float64 datatype, columns are: Estimated Unemployment Rate (%), Estimated Employed, Estimated Labour Participation Rate (%).

```python
# Checking the null values in the dataset:

df2.isnull().sum()
```

```python
print('The shape of the dataset before removing "NAN" values',df2.shape)
```

```python
dff = df2.dropna()
```

```python
print('The shape of the dataset after removing "NAN" values',dff.shape)
```

```python
dff.head()
```

```python
dff.columns
```

```python
dff.info()
```

```python
print("Unemployment Data are grouped by:",dff.groupby('Region').size())
```

## 4. MULTIVARIATE ANALYSIS:

```python
dff.corr()
```

```python
plt.figure(figsize=(10,5))

sns.heatmap(dff.corr(),linewidths='.5',fmt='.1f',cmap='Spectral')

plt.show()
```

```python
sns.pairplot(dff, hue="Region", palette='Spectral_r');
```

```python
reg_freq = dff['Region'].value_counts()
```

```python
reg_freq.plot(kind='pie', startangle = 0)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,15))

sns.countplot(x='Region',data=dff, orient='h')
plt.xticks(rotation=45)
plt.ylabel('Estimated Unemployment Rate (%)')
```
